## QLearning on TicTacToe

```python
from pprint import pprint

import gym
from stable_baselines3 import PPO
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.vec_env import DummyVecEnv, VecVideoRecorder
import wandb
from wandb.integration.sb3 import WandbCallback

from rl_agents.temporal_difference.tabular_agents import QLearning, TabularQFunction
from environments.tictactoe import TicTacToe

config = {
    "algorithm": "QLearning",
    "policy": "TabularQFunction",
    "env": {"opponents": ["random", "reaction"],
            "switching_prob": 0.5},
}

# run = wandb.init(
#     project="tictactoe",
#     config=config,
#     sync_tensorboard=True,  # auto-upload sb3's tensorboard metrics
#     monitor_gym=False,  # auto-upload the videos of agents playing the game
#     save_code=False,  # optional
# )

def make_env():
    env = TicTacToe(opponents=config["env"]["opponents"],
                    switching_prob=config["env"]["switching_prob"])
    env = Monitor(env)  # record stats such as returns
    return env

env = DummyVecEnv([make_env])

q_learning = QLearning(policy=TabularQFunction,
                       env=env, 
                       verbose=1,
                       tensorboard_log=f"logs/{config['algorithm']}_tictactoe"
                      )
```

```python
q_learning.learn(total_timesteps=5000,
                 log_interval=10,
                 selfplay_start=50,
                 reset_num_timesteps=False,
#                  callback=WandbCallback(log="all")
                )
```

## MCTS on TicTacToe

```python
from pprint import pprint

import gym
from stable_baselines3 import PPO
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.vec_env import DummyVecEnv, VecVideoRecorder
import wandb
from wandb.integration.sb3 import WandbCallback

from rl_agents.monte_carlo_tree_search.mcts_agent import MCTS, MCTSPolicy
from environments.tictactoe import TicTacToe


config = {
    "algorithm": "MCTS",
    "policy": "MCTSPolicy",
    "env": {"opponents": ["random", "reaction"],
            "switching_prob": 0.5},
    "total_timesteps": 10000,
    "log_interval": 100,
    "selfplay_start": 1000
}

run = wandb.init(
    project="tictactoe_final",
    config=config,
    sync_tensorboard=False,  # auto-upload sb3's tensorboard metrics
    monitor_gym=False,  # auto-upload the videos of agents playing the game
    save_code=False,  # optional
)

def make_env():
    env = TicTacToe(opponents=config["env"]["opponents"],
                    switching_prob=config["env"]["switching_prob"])
    env = Monitor(env)  # record stats such as returns
    return env

env = DummyVecEnv([make_env])

mcts_agent = MCTS(policy=MCTSPolicy,
                  env=env,
                  max_episode_steps=10,
                  verbose=1,
                  tensorboard_log=f"logs/{config['algorithm']}_tictactoe"
                 )
```

```python
mcts_agent.learn(total_timesteps=config["total_timesteps"],
                 log_interval=config["log_interval"],
                 selfplay_start=config["selfplay_start"],
                 reset_num_timesteps=False,
                 callback=WandbCallback(log="all",
                                        model_save_path=f"pretrained_agents/{run.name}",
                                       )
                )
```

## WandB demo code

```python
import gym
from stable_baselines3 import PPO
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.vec_env import DummyVecEnv, VecVideoRecorder
import wandb
from wandb.integration.sb3 import WandbCallback

from environments.tictactoe import TicTacToe

config = {
    "policy_type": "MultiInputPolicy",
    "total_timesteps": 25000,
    "env_name": "CartPole-v1",
}
run = wandb.init(
    project="tictactoe_final",
    config=config,
    sync_tensorboard=False,  # auto-upload sb3's tensorboard metrics
    monitor_gym=False,  # auto-upload the videos of agents playing the game
    save_code=False,  # optional
)


def make_env():
    env = TicTacToe(opponents=["reaction"], switching_prob=0.5)
    env = Monitor(env)  # record stats such as returns
    return env


env = DummyVecEnv([make_env])

model = PPO(config["policy_type"], env, verbose=1, tensorboard_log=f"runs")
model.learn(
    total_timesteps=config["total_timesteps"],
    callback=WandbCallback(
#         model_save_path=f"models/{run.id}",
#         gradient_save_freq=100,
        log="all",
    ),
)
```
